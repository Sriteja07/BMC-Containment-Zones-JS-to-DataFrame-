```
#reading javascript file
js = open("BMC-ContainmentZones (1).js", "r").readlines()
```

```
print('Total lines present in javascript file : ',len(js))
print(js[0])
print(js[len(js)-1])
print(type(js))
print(type(js[87]))
```

```
#Finding markers , latitudes and longitudes using Regular Expressions library
import re
from tqdm import tqdm
pattern = re.compile(r'var marker') 
marker_list = []
count = 0
for line in tqdm(js): 
  temp = []
  sr =  pattern.search(line) == None 
  if ((sr == False) and (count<(len(js)-1))):
    temp = str(line) + str(js[count+1])
    marker_list.append(temp)
  count= count +1 
```

```
marker_list[0].split()
```

```
markers = []
latitude = []
longitude = []
for i in marker_list:
  markers.append(i.split()[1])
  latitude.append(i.split()[4])
  longitude.append(i.split()[5])
```

```
for i in range(0,len(markers)):
  markers[i] = markers[i].replace('marker_', '')
  latitude[i] = latitude[i].replace('[','')
  latitude[i] = latitude[i].replace(',','')
  longitude[i] = longitude[i].replace(']','')
  longitude[i] = longitude[i].replace(',','')
```

```
import pandas as pd
Geo_data = pd.DataFrame()
Geo_data['marker_id'] = markers
Geo_data['latitude'] = latitude
Geo_data['longitude'] = longitude
Geo_data.head()
```
